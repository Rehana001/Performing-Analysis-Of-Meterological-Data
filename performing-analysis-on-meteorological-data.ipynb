### resampling appearent temperature and humidity by resample function to plot graph against them 

```python
df=pd.read_csv('../input/d/rehananoorani/weather-dataset/weather_data.csv')
titles_req=['Formatted Date', 'Apparent Temperature (C)','Humidity','Summary']
df_1=df[titles_req]
df_1.rename({'Apparent Temperature (C)':'Apparent Temperature'},axis=1,inplace=True)
df_1['Formatted Date']=pd.to_datetime(df_1['Formatted Date'],utc=True)
df_2=df_1.set_index('Formatted Date')
df_2.resample('M').mean()
```

### lets import seaborn library and plot graphs using matplotlib 

```python
import seaborn as sns 
sns.pairplot(df_2,kind='scatter')
plt.show()
```

### variation of temperature and humidity along with time 

```python
plt.figure(figsize=(20,20))
sns.pairplot(df_2,hue="Summary")
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.relplot(x='Apparent Temperature',y='Humidity',color='purple',hue="Summary",data=df_2)
```

### Yearly variation of temperature vs humidity 

```python
plt.figure(figsize=(20,10))
plt.title("Relation of Apparent Temperature and Humidity")
sns.lineplot(x="Apparent Temperature",y="Humidity",data=df_2 ,color='green', linestyle='dashed')
```

### Plot of Apparent Temperature and Humidity according to year
